## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predictive,' etc.)*

**What this work CAN claim:**
- ✅ **Observed patterns:** "Pages in category X tend to have lower engagement than category Y"
- ✅ **Directional insights:** "Pages with downward trends are more likely to benefit from review"
- ✅ **Decision support:** "This ranking prioritizes pages for human review based on historical performance signals"
- ✅ **Correlation:** "Pages with older content updates show lower average sessions"

**What this work CANNOT claim:**
- ❌ **Causal proof:** "Refreshing this page WILL increase traffic" (correlation ≠ causation)
- ❌ **Predictive guarantees:** "This page will definitely improve if updated" (we don't know implementation quality)
- ❌ **Specific ROI numbers:** We can't promise "this refresh will increase traffic by 40%"
- ❌ **Universal truth:** Different content types, industries, and seasons behave differently

**Bottom line:** This is a **triage and prioritization tool**, not a fortune teller.
